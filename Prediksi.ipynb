```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Buat dataset
data = {
    "No": list(range(1, 21)),
    "Tahun": list(range(2005, 2025)),
    "Jumlah Iklan": [2.0, 2.5, 3.0, 3.5, 4.0, 4.2, 4.5, 4.7, 5.0, 5.5,
                     6.0, 6.2, 6.5, 6.7, 7.0, 7.5, 8.0, 8.5, 9.0, 9.5],
    "Penjualan": [20, 22, 25, 27, 29, 30, 32, 33, 35, 38,
                  40, 42, 44, 45, 47, 49, 52, 54, 56, 58]
}

df = pd.DataFrame(data)

# Regresi Linear
X = df["Jumlah Iklan"].values
Y = df["Penjualan"].values

# Tambahkan konstanta untuk intercept
X_matrix = np.vstack([np.ones(len(X)), X]).T
beta = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ Y
intercept, slope = beta

# Prediksi 2025 dan 2026
prediksi_2025 = intercept + slope * 10.0
prediksi_2026 = intercept + slope * 10.5

# Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color="blue", label="Data Asli")
plt.plot(X, intercept + slope * X, color="red", label=f"Regresi Linear: y = {intercept:.2f} + {slope:.2f}x")
plt.scatter([10.0, 10.5], [prediksi_2025, prediksi_2026], color="green", label="Prediksi 2025 & 2026")
plt.title("Regresi Linear: Jumlah Iklan vs Penjualan")
plt.xlabel("Jumlah Iklan (juta)")
plt.ylabel("Penjualan (juta)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Simpan file
excel_path = "regresi_iklan_penjualan.xlsx"
chart_path = "regresi_iklan_penjualan.png"
df.to_excel(excel_path, index=False)
plt.savefig(chart_path)

excel_path, chart_path
```
